# <span style='color:teal;font-size:32px;font-family: "Times New Roman", Times, serif'>Graphs in Data Visualization </span>
# <span style='color:purple;font-size:25px;font-family: "Times New Roman", Times, serif'>1.Line Graph</span>
# <span style='color:mediumpurple;font-size:20px;font-family: "Times New Roman", Times, serif'>1.1 Single Line Graph</span>
# <span style='color:mediumseagreen;font-size:18px;font-family: "Times New Roman", Times, serif'>A line graph (also known as a line plot or line chart) is a graph which uses lines to connect individual data points that display quantitative values over a specified time interval. Line graphs use data point "markers" that are connected by straight lines to aid in visualization.</span>
# <span style='color:mediumseagreen;font-size:18px;font-family: "Times New Roman", Times, serif'> It is used when:Line graphs are used to track changes over short and long periods of time. When smaller changes exist, line graphs are better to use than bar graphs. Line graphs can also be used to compare changes over the same period of time for more than one group</span>
  

```python
x=data1['Year']
y=data1['Total']
plt.figure(figsize=(8,8))                                                         #figure function is used to set size
plt.plot(data1['Year'],data1['Total'])                                            #for plotting the graph
plt.title("Rate of literacy from 1996 to 2016")                                   #title of the graph
plt.legend(labels=['Percentage'])                                                 #legends of the graph
plt.xlabel("Year")                                                                #label on X-axis
plt.ylabel("Percentage of literacy ")                                             #label on Y-axis
plt.plot(x,y,color='blue',linewidth=2,marker='o',markersize=6,markeredgecolor='springgreen')  #for better visualization effect
plt.show()                                                                        # show function used to show developed plot 
```

# From the above line graph we can conclude the rate of literacy has increased over the years.

# <span style='color:mediumpurple;font-size:20px;font-family: "Times New Roman", Times, serif'> 1.2 Multiple Line Graph</span>

```python
x=data1['Year']
y1=data1['LRFP']
y2=data1['LRMP']
plt.figure(figsize=(10,10))
plt.rcParams['axes.facecolor'] = 'black'                                             #background of the graph
plt.plot(x,y1,'r',linewidth=3.5,marker='o',markersize=5.5,markeredgecolor='white')   #to make graph more attractive and 
plt.plot(x,y2,'g',linewidth=3.5,marker='o',markersize=5.5,markeredgecolor='white')   #to increase readability
plt.title("Gender Wise Percentage")
plt.legend(labels=["Female","Male"])                                                 #legends of graph
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.show()
```

# From the above graph we can conclude that female percentage of literacy is pretty much less as compared to male percentage of literacy.

# <span style='color:purple;font-size:25px;font-family: "Times New Roman", Times, serif'> 5.Histogram</span>
# <span style='color:mediumseagreen;font-size:18px;font-family: "Times New Roman", Times, serif'> Quality Glossary Definition: Histogram. A frequency distribution shows how often each different value in a set of data occurs. A histogram is the most commonly used graph to show frequency distributions. It looks very much like a bar chart, but there are important differences between them.</span>
# <span style='color:mediumseagreen;font-size:18px;font-family: "Times New Roman", Times, serif'> A frequency distribution shows how often each different value in a set of data occurs. A histogram is the most commonly used graph to show frequency distributions. It looks very much like a bar chart, but there are important differences between them.</span>

```python
plt.figure(figsize=(8,8))
plt.rcParams['axes.facecolor'] = 'lightcyan' 
# plot done using matplotlib
plt.hist(data1["Total"],bins=10,facecolor='coral')
mean=data1["Total"].mean()
median=data1["Total"].median()
plt.axvline(mean,color="red",linewidth=5)
plt.axvline(median,color="lime",linewidth=5)
plt.title("Literacy rate ")
plt.ylabel("frequency")
plt.xlabel("Literacy Rate")
plt.show()
```
